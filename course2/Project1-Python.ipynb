# Project 1 : Average Temperature of City

```python
import numpy as np
import pandas as pd
```

Import dataframe

```python
data = pd.read_csv("files/citytemp.csv")
data
```

### Beri nama kolom pada data

```python
main_data = data
main_data.columns = ["City", "Temperature", "Unit"]
main_data
```

### Rubah dari celcius ke fahrenheit

```python
converted = main_data.copy()
id = converted[converted["Unit"] == "C"].index
converted.loc[id, "Temperature"] = (converted.loc[id, "Temperature"] * 9/5) + 32
converted = converted.drop(columns="Unit")
converted
```

### Buat variable untuk menampung nama-nama kota

```python
# hapus baris yang memiliki kesamaan nama kota dan masukkan data ke cities
# cities = main_data.drop_duplicates(subset=["City"], keep="first")["City"]
# cities

# menggunakan groupby
average = converted.groupby(converted["City"]).mean(numeric_only=True).round(3)
average.rename(columns={"Temperature" : "Average Temp (F)"}, inplace=True)
average
```
